```python
import pandas as pd
import numpy as np
```

```python
#Choose a league, season, and 4 teams within the league. For the given season, collect the roster table from sports reference for ALL 4 teams. This should result in you having 4 different CSVs(one for each team)
# I pulled World Cup roster from the 4 Conacaf Teams in 2022 World Cup- USA, Mexico, Costa Rica, and Canada 

#In your notebook file, read all 4 CSVs in, with the index being the player name.
mexico_worldcuproster=pd.read_csv("data/mexicowcup22.csv",index_col='Player')
canada_worldcuproster=pd.read_csv("data/canadawcup22.csv",index_col='Player')
costarica_worldcuproster=pd.read_csv("data/costaricawcup22.csv",index_col='Player')
usa_worldcuproster=pd.read_csv("data/usmntwcup22.csv",index_col='Player')
mexico_worldcuproster
canada_worldcuproster
costarica_worldcuproster
usa_worldcuproster


```

```python
#Concatenate the 4 tables into 1 all-inclusive table.
concacaf_worldcupteams =pd.concat([mexico_worldcuproster,canada_worldcuproster,usa_worldcuproster,costarica_worldcuproster])
concacaf_worldcupteams
```

```python
#Write the concatenated table to CSV. (Note: the index SHOULD be included)
concacaf_worldcupteams.to_csv('concacaf_worldcupteams.csv')
```
